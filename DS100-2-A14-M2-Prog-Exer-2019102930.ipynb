```python
import pandas as pd
import numpy as np
import matplotlib.animation as animation
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_row',None)
data=pd.read_csv('https://github.com/KevinPuatu/DS100/blob/6bd1bb11df207d0621f1b273bc1a52cbb4fbc117/F1DriversDataset.csv?raw=true')
data
```

```python
%matplotlib notebook
#Create a Syntax that excludes the non-race winners from United Kingdom
def_UnitedKingdom=data.loc[(data)['Nationality'].str.contains('United Kingdom')&data['Race_Wins']!=0]
#Define the Syntax to use
UnitedKingdom = def_UnitedKingdom[['Driver','Race_Wins']]
#Plot the Bar Chart
graph=UnitedKingdom.plot.bar ('Driver','Race_Wins', rot = 20, color='blue', width=1) #rot must be used to display the names clearly
plt.title('United Kingdom Drivers with Respect to their # of Race Wins')
#Adjusting the Plotting Points
plt.xticks(fontsize=5)
plt.yticks(fontsize=10)
plt.xlabel('Drivers')
plt.ylabel('Race Wins')
plt.grid(True) #To enable the grid lines to further easily observe the data
plt.show()
```

```python
#Create a Syntax that excludes the non-race winners from Germany
def_Germany=data.loc[(data)['Nationality'].str.contains('Germany')&data['Race_Wins']!=0]
#Define the Syntax
Germany=def_Germany[['Driver','Race_Wins']]
#Plot the Line Graph
graph=Germany.plot.line('Driver','Race_Wins', rot=10, color='Orange') #rot is used to rotate the driver's name to read clearly
#Adjusting the Plotting Points
plt.title('German Drivers with Respect to their # of Race Wins')
plt.xticks(fontsize=5.5)
plt.yticks(fontsize=10)
plt.xlabel('Drivers')
plt.ylabel('Race Wins')
plt.grid(True) #To enable the grid lines to further easily observe the data
plt.show()
```

```python
#Create a Syntax that excludes the non-championship winners from United Kingdom
def_UnitedKingdom=data.loc[(data)['Nationality'].str.contains('United Kingdom')&data['Championships']!=0]
#Define the Syntax
UnitedKingdom=def_UnitedKingdom[['Driver','Championships']]
#Plot the Scatter Plot
graph=UnitedKingdom.plot.scatter('Driver','Championships',rot=10,color='red') #rot must be used to display the drivers' name clearly
#Adjusting the Plotting Points
plt.title('United Kingdom Drivers World Champions')
plt.xticks(fontsize=6)
plt.yticks(fontsize=10)
plt.xlabel('Drivers')
plt.ylabel('Championships')
plt.grid(True) #To enable grid lines
plt.show()
```

```python
#Create a Syntax that excludes the non-race winners from Germany
def_Germany=data.loc[(data)['Nationality'].str.contains('Germany')&data['Race_Wins']!=0]
#Define the Syntax
Germany=def_Germany['Race_Wins']
#Create a new figure to avoid overlapping with the figure above
plt.figure()
#Plot the Histogram
plt.title('Histogram of German Drivers Frequency of Race Wins')
plt.hist(Germany, bins=30, color='Pink')
plt.xlabel('Race Wins')
plt.ylabel('Frequency')
plt.grid(True) #To enable grid lines
plt.show()
```
